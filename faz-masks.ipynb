```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
MARKED_PATH='/kaggle/input/fazmasks/FAZmasks/marked'
OUTPUT_PATH='/kaggle/working/masks'
show_sample=True
```

```python
os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
lower_yellow = np.array([20,100,100])
upper_yellow= np.array([35,255,255])
```

```python
def extract_faz_mask(image_path):
    img=cv2.imread(image_path)
    hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #convert from BGR to HSV colour
    mask=cv2.inRange(hsv, lower_yellow, upper_yellow) #taking only yellow area
    #to clear noise
    kernel=np.ones((3,3),np.uint8)
    mask=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
    return mask
```

```python
generated_count=0
for file in sorted (os.listdir(marked_path)):
    if file.lower().endswith(('.jpg','.png')) and 'marked' in file.lower():
        input_path=os.path.join(marked_path, file)
        mask=extract_faz_mask(input_path)
        base_name=file.replace('_marked','').split('.')[0]
        output_file=f"{base_name}_mask.png"
        output_path=os.path.join(OUTPUT_PATH, output_file)
        cv2.imwrite(output_path, mask)
        generated_count+=1
        print(f"saved:{output_file}")
print(f"\n total masks generated: {generated_count}")
```
